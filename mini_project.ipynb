```python
crime_list = ['살인', '강도', '방화', '성폭력']
for crime in crime_list :
    if crime == '살인' :
        condi1 = ((crime_area_df['범죄별(3)']==f'{crime}'))
        df = crime_area_df[condi1]
    if crime == '강도' :
        condi2 = (crime_area_df['범죄별(3)']==f'{crime}')
        df = df.append(crime_area_df[condi2])
    if crime == '방화' :
        condi3 = (crime_area_df['범죄별(3)']==f'{crime}')
        df = df.append(crime_area_df[condi3])
    if crime == '성폭력' :
        condi4 = (crime_area_df['범죄별(3)']==f'{crime}')
        df = df.append(crime_area_df[condi4])
```

```python
location_data = library[['위도', '경도']].values[:len(library)].tolist()
```

```python
x=folium.Map(location=[37.541,126.986],tiles='Stamen Toner',zoom_start=10)
for i in range(len(df_test)) :
    for j in range(len(df_test2.columns)) :
        if df_test['범죄별'].tolist()[i] == '살인' :
            df_test_crime = '살인'
            crime_color = 'blue'
            max_radius = int(df_test2[df_test2.columns[j]].tolist()[i])        
            folium.Circle(location=location_data[j], radius=max_radius*300, color=crime_color,
                          fill=True, fill_opacity=1, tooltip=df_test_crime).add_to(x)         
            
        elif df_test['범죄별'].tolist()[i] == '강도' :
            df_test_crime = '강도'
            crime_color = 'white'
            max_radius = int(df_test2[df_test2.columns[j]].tolist()[i])        
            folium.Circle(location=location_data[j], radius=max_radius*300, color=crime_color,
                          fill=True, fill_opacity=0.7, tooltip=df_test_crime).add_to(x)
            
        elif df_test['범죄별'].tolist()[i] == '방화' :
            df_test_crime = '방화'
            crime_color = 'red'
            max_radius = int(df_test2[df_test2.columns[j]].tolist()[i])        
            folium.Circle(location=location_data[j], radius=max_radius*300, color=crime_color,
                          fill=True, fill_opacity=0.7, tooltip=df_test_crime).add_to(x)
        elif df_test['범죄별'].tolist()[i] == '성폭력' :
            df_test_crime = '성폭력'
            crime_color = 'yellow'
            max_radius = int(df_test2[df_test2.columns[j]].tolist()[i])        
            folium.Circle(location=location_data[j], radius=max_radius*3, color=crime_color,
                          fill=True, fill_opacity=0.5, tooltip=df_test_crime).add_to(x)

plugins.HeatMap(location_data).add_to(x)
```
